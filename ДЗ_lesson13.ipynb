ЗАДАНИЯ С API:
1. Количество капсул каждого типа
2. Среднее время между первым и последним запуском
3. Капсулы без информации о запуске
4. Сравнение Dragon 1.0 и Dragon 1.1. Среднее кол-во повторного использования, среднее кол-во посадок

```python
import requests

responce = requests.get('https://api.spacexdata.com/v3/capsules')

data = responce.json()
df = pd.DataFrame(data)
print(df['type'].value_counts())  #метод в Pandas для подсчета уникальных значений в Series
```

```python

df['date'] = pd.to_datetime(df['original_launch'], errors='coerce')
df['date'].mean()
time_difference = df['date'].max() - df['date'].min() # Вычисляем разницу между последним и первым запусками
num_launches = df['date'].count()                     # Считаем количество запусков
average_time = time_difference / (num_launches - 1)   # Вычитаем 1 так как количество промежутков между запусками всегда на 1 меньше общего числа запусков
print("Среднее время между первым и последним запуском:", average_time)
```
